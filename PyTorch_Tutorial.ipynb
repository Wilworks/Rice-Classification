```python
import torch
import torch.nn as nn

# Set device
device = torch.device("mps")
print(f"Using device: {device}")

# Define a simple model
class SimpleModel(nn.Module):
    def __init__(self):
        super(SimpleModel, self).__init__()
        self.fc1 = nn.Linear(784, 128)
        self.fc2 = nn.Linear(128, 10)
        self.relu = nn.ReLU()
    
    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# Create model and move to GPU
model = SimpleModel().to(device)

# Create some dummy data and move to GPU
data = torch.randn(32, 784).to(device)
labels = torch.randint(0, 10, (32,)).to(device)

# Test forward pass
output = model(data)
print(f"Output shape: {output.shape}")
print("Model is running on MPS!")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import torch.nn.functional as F
```

```python
Df = pd.read_csv("Iris.csv")
Df.head()
```

```python
#Species Encoding
species_to_idx = {
    "Iris-setosa": 0,
    "Iris-versicolor": 1,
    "Iris-virginica": 2
}
Df["Species"] = Df["Species"].map(species_to_idx)
Df.head()
```

```python
X = Df.drop("Species", axis=1)
y = Df["Species"]
#Convert to numpy arrays
X = X.values
y = y.values
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=41)
```

```python

# Move data to device
X_train = X_train.to(device)
y_train = y_train.to(device)

# Setup optimizer and criterion
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.00001)

# Now train!
epochs = 100000
for i in range(epochs):
    y_pred = model(X_train)  # You can use model(X) instead of model.forward(X)
    loss = criterion(y_pred, y_train)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    if i % 10 == 0:
        print(f'Epoch {i}, Loss: {loss.item():.4f}')
```

```python
# Move to CPU and convert to numpy
loss_cpu = loss.cpu().detach().numpy()
plt.plot(loss_cpu)
plt.show()
```

```python
#Evaluate the model
#Removes Backpropagation
with torch.no_grad():
    y_eval  = model.forward(X_test.to(device))
    loss = criterion(y_eval, y_test.to(device))
    print(f"Evaluation Loss: {loss.item():.4f}") == criterion(y_eval, y_test.to(device))
    print(f"Evaluation Loss: {loss.item():.4f}")
```

```python
#Performance Metrics
_, predictions = torch.max(y_eval, 1)
correct = (predictions == y_test.to(device)).sum().item()
accuracy = correct / y_test.size(0)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
